Standard Scalar

```python
scaled_features = data.copy()
```

```python
featuresX = scaled_features[["X"]]
standard_scaler = preprocessing.StandardScaler()
scaled_features["X"] = standard_scaler.fit_transform(featuresX.values)
```

```python
scaled_features["X"]
```

```python
featuresY = scaled_features[["Y"]]
standard_scaler = preprocessing.StandardScaler()
scaled_features["Y"] = standard_scaler.fit_transform(featuresY.values)
```

```python
test2 = pd.DataFrame(data = scaled_features,columns=["X","Y"])
sns.boxplot(x="variable",y="value",data=pd.melt(test2),palette="Set1")
plt.show()
```

MinMax Scalar

```python
MinMax_features = data.copy()
featuresX = MinMax_features[["X"]]
MinMax_scaler = preprocessing.MinMaxScaler()
MinMax_features["X"] = MinMax_scaler.fit_transform(featuresX.values)
featuresY = MinMax_features[["Y"]]
MinMax_scaler = preprocessing.MinMaxScaler()
MinMax_features["Y"] = MinMax_scaler.fit_transform(featuresY.values)
```

```python
test3 = pd.DataFrame(data = MinMax_features,columns=["X","Y"])
sns.boxplot(x="variable",y="value",data=pd.melt(test3),palette="Set1")
plt.show()
```
